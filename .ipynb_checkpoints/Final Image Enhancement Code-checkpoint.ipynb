# Object Detection using Haar Cascade

```python
#read image

img = cv2.imread(r'C:\Users\gayathri_sri_mamidib\Desktop\Hackathons\Eye Doctor\Sample\test_girl.png')

#convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#apply haar cascade
haar_face = cv2.CascadeClassifier(r'C:\Users\gayathri_sri_mamidib\Desktop\Hackathons\Eye Doctor\Sample\model\haarcascade_frontalface_default.xml')
haar_eyes = cv2.CascadeClassifier(r'C:\Users\gayathri_sri_mamidib\Desktop\Hackathons\Eye Doctor\Sample\model\haarcascade_eye.xml')
faces = haar_face.detectMultiScale(gray,1.3,5)
```

```python
print(faces) # order - x-axis, y-axis, width, height
```

```python
for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
    #crop the face
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w] 
plt.imshow(img) #color image in BGR format
```
